```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import re
```

```python
states_list = ["Alaska", "Alabama", "Arkansas", "Arizona", "California", "Colorado", "Connecticut", "District of Columbia", "Delaware", "Florida", "Georgia", "Guam", "Hawaii", "Iowa", "Idaho", "Illinois", "Indiana", "Kansas", "Kentucky", "Louisiana", "Massachusetts", "Maryland", "Maine", "Michigan", "Minnesota", "Missouri", "Mississippi", "Montana", "North Carolina", "North Dakota", "Nebraska", "New Hampshire", "New Jersey", "New Mexico", "Nevada", "New York", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Puerto Rico", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Virginia", "Virgin Islands", "Vermont", "Washington", "Wisconsin", "West Virginia", "Wyoming"]
states_list2 = []
# lower, and also replace spaces with - 
regex = re.compile('[%s]' % re.escape("[-,\.!?:â€™ ]"))
for state in states_list:
    state_lower = state.lower()
    states_list2.append(regex.sub('-', state_lower))
states_list2
```

```python
# We can now automate the process of web scrapping
# lists as necessary 
county_list = []
state_list = []
br0 = []
br1 = []
br2 = []
br3 = []
br4 = []
state_ind = 0
for state in states_list2:
    print(state)
    if state not in ['district-of-columbia', 'guam']:
    # Format the url, obtain soup object
        web_url = 'https://www.rentdata.org/states/{}/2015'.format(state)
        print(web_url)
        rent_res = requests.get(web_url)
        soup_rent = BeautifulSoup(rent_res.content)
        rent_table = soup_rent.find_all('table', class_ = 'hover scroll tablesorter')
        # Obtain the county names
        county_table = soup_rent.find_all('b')
        state_name = county_table[0].text.strip()
        # print(state_name)
        county_table = county_table[3:]
        for item in county_table:
            county_list.append(item.text.strip())
            state_list.append(state_name)
        # print(county_list)
        cost_table = soup_rent.find_all('span', class_ = 'money')
        cost_table = cost_table[5:]
        # Turn costs into floats
        cost_list1 = []
        cost_list2 = []
        for item in cost_table:
            cost_list1.append(item.text.split()[0])
        
        # remove $, then turn into float 
        for item in cost_list1:
            money_text = regex.sub("", item.split('$')[1])
            money_float = float(money_text)
            cost_list2.append(money_float)
        # iterate through the cost lists to get the rents across counties
        ind0 = 0
        ind1 = 1
        ind2 = 2
        ind3 = 3
        ind4 = 4
        while ind0 in range(len(cost_list2)):
            br0.append(cost_list2[ind0])
            br1.append(cost_list2[ind1])
            br2.append(cost_list2[ind2])
            br3.append(cost_list2[ind3])
            # print(cost_list2[ind3])
            br4.append(cost_list2[ind4])
            ind0 += 5
            ind1 += 5
            ind2 += 5
            ind3 += 5
            ind4 += 5
    else:
        # Format the url, obtain soup object
        web_url = 'https://www.rentdata.org/states/{}/2015'.format(state)
        print(web_url)
        rent_res = requests.get(web_url)
        soup_rent = BeautifulSoup(rent_res.content)
        rent_table = soup_rent.find_all('table', class_ = 'hover scroll tablesorter')
        # Obtain the county names
        county_table = soup_rent.find_all('b')
        state_name = county_table[0].text.strip()
        county_table = county_table[3:]
        for item in county_table:
            county_list.append(item.text.strip())
            state_list.append(state_name)
        # print(county_list)
        cost_table = soup_rent.find_all('span', class_ = 'money')
        cost_table = cost_table[4:]
        # Turn costs into floats
        cost_list1 = []
        cost_list2 = []
        for item in cost_table:
            cost_list1.append(item.text.split()[0])
        # remove $, then turn into float 
        for item in cost_list1:
            money_text = regex.sub("", item.split('$')[1])
            money_float = float(money_text)
            cost_list2.append(money_float)
        # iterate through the cost lists to get the rents across counties
        ind0 = 0
        ind1 = 1
        ind2 = 2
        ind3 = 3
        ind4 = 4
        while ind0 in range(len(cost_list2)):
            br0.append(cost_list2[ind0])
            print(cost_list2[ind0])
            br1.append(cost_list2[ind1])
            br2.append(cost_list2[ind2])
            br3.append(cost_list2[ind3])
            print(cost_list2[ind3])
            br4.append(cost_list2[ind4])
            ind0 += 5
            ind1 += 5
            ind2 += 5
            ind3 += 5
            ind4 += 5
rent_df_county = pd.DataFrame(list(zip(state_list, county_list, br0, br1, br2, br3, br4)), columns = ['State Name', 'County Name', 'Rent for 0 Bedrooms',
                                                                                     'Rent for 1 Bedroom', 'Rent for 2 Bedrooms',
                                                                                     'Rent for 3 Bedrooms', 'Rent for 4 Bedrooms'])
rent_df_county.to_csv(r'C:\Users\huang\OneDrive\Documents\ECO225\2015_US_County_Rent.csv')
```
